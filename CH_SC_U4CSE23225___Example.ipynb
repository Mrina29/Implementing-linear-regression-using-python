```python
### COMPLETE, GUARANTEED SOLUTION FOR SALARY PREDICTION ###

# ==============================================================================
# Step 1: Import all required libraries
# ==============================================================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# ==============================================================================
# Step 2: Create the Dataset Directly in Code (No Download Needed)
# ==============================================================================
# This data is the exact content of the original 'Salary_Data.csv' file.
# This method is guaranteed to work because it has no external dependencies.
salary_data_dict = {
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5],
    'Salary': [39343.0, 46205.0, 37731.0, 43525.0, 39891.0, 56642.0, 60150.0, 54445.0, 64445.0, 57189.0, 63218.0, 55794.0, 56957.0, 57081.0, 61111.0, 67938.0, 66029.0, 83088.0, 81363.0, 93940.0, 91738.0, 98273.0, 101302.0, 113812.0, 109431.0, 105582.0, 116969.0, 112635.0, 122391.0, 121872.0]
}

data = pd.DataFrame(salary_data_dict)
print("✅ Dataset created successfully! No download was needed.")

# ==============================================================================
# Step 3: Explore the Dataset
# ==============================================================================
print("\nFirst 5 rows of the dataset:")
print(data.head())

# ==============================================================================
# Step 4: Define Features (X) and Target (y), then Split
# ==============================================================================
X = data[['YearsExperience']]
y = data['Salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nData split into {len(X_train)} training samples and {len(X_test)} testing samples.")

# ==============================================================================
# Step 5: Create and Train the Linear Regression Model
# ==============================================================================
model = LinearRegression()
model.fit(X_train, y_train)
print("✅ Model trained successfully!")

# ==============================================================================
# Step 6: Make Predictions on the Test Data
# ==============================================================================
y_pred = model.predict(X_test)

# ==============================================================================
# Step 7: Visualize the Results
# ==============================================================================
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='skyblue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Actual Test Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Regression Line')
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
plt.legend()
plt.grid(True)
plt.show()

print("\n✅ This completes the exercise!")
```
